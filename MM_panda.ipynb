```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("/mortalidad_materna_2002_2022.xlsx")

print(df.columns)

print(df.head())

print(df.info())


plt.figure(figsize=(10, 6))

df['EDAD_QUINQUENALD'].value_counts().sort_index().plot(kind='bar')
plt.title('Muertes Maternas por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Muertes')
plt.grid(axis='y')
plt.show()
```
